## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


# 📝 **กิจกรรมในแลป**

1. **แบบฝึกหัด 1**: ใช้ NumPy คำนวณค่า **Mean, Max, Min** ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง
2. **แบบฝึกหัด 2**: ใช้ GeoPandas โหลด **Shapefile** ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด
3. **แบบฝึกหัด 3**: ใช้ GeoPandas ทำ **Spatial Join** ระหว่างข้อมูลจังหวัดและอำเภอ
4. **แบบฝึกหัด 4**: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด


```python
import numpy as np

# 1. สร้างอาร์เรย์สมมติขนาด 5x5 ที่มีค่าระหว่าง -1.0 ถึง 1.0
# เราจะใช้ np.random.uniform เพื่อจำลองค่า NDVI
ndvi_array = np.random.uniform(-1, 1, (5, 5))

print("--- ข้อมูล NDVI สมมติ (5x5) ---")
print(ndvi_array)
print("-" * 30)

# 2. คำนวณค่าทางสถิติโดยใช้ NumPy
mean_ndvi = np.mean(ndvi_array)
max_ndvi  = np.max(ndvi_array)
min_ndvi  = np.min(ndvi_array)

# 3. แสดงผลลัพธ์
print(f"ค่าเฉลี่ย (Mean): {mean_ndvi:.4f}")
print(f"ค่าสูงสุด (Max):  {max_ndvi:.4f}")
print(f"ค่าต่ำสุด (Min):  {min_ndvi:.4f}")

# 4. แถม: นับจำนวน Pixel ที่เป็นพื้นที่สีเขียว (สมมติว่า NDVI > 0.5)
green_pixels = np.count_nonzero(ndvi_array > 0.5)
print(f"จำนวน Pixel ที่เป็นป่า/พืชพรรณหนาแน่น (> 0.5): {green_pixels} จุด")
```

**np.random.uniform(-1, 1, (5, 5))**: สร้างเลขสุ่มแบบทศนิยมในช่วง -1 ถึง 1 ขนาด 5 แถว 5 คอลัมน์

**np.mean()**: หาค่าเฉลี่ยของข้อมูลทั้งหมดในอาร์เรย์

**np.max() / np.min()**: หาค่าที่มากที่สุดและน้อยที่สุด

**ndvi_array > 0.5**: เป็นการสร้าง Mask (เงื่อนไข) เพื่อกรองข้อมูลที่สนใจ

```python
import geopandas as gpd

# โหลดไฟล์ (แก้ path ตามจริง)
gdf_provinces = gpd.read_file("/content/drive/MyDrive/GE234/tha_admbnda_adm1_rtsd_20190221.gpkg")

# แปลงเป็น EPSG:32647 (UTM Zone 47N สำหรับไทย) เพื่อหาพื้นที่เป็นตารางเมตร
gdf_provinces['area_sqkm'] = gdf_provinces.to_crs(epsg=32647).area / 10**6

print(gdf_provinces[['ADM1_TH', 'area_sqkm']].head())
```

**จุดสำคัญ**
ต้องแปลง CRS เป็นหน่วยเมตร (เช่น UTM) ก่อนคำนวณพื้นที่ ไม่เช่นนั้นจะได้ค่าเป็น Decimal Degrees

```python
# โหลดข้อมูลอำเภอ
gdf_amphoe = gpd.read_file("/content/drive/MyDrive/GE234/THA_Amphoe.shp")

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf_provinces = gpd.read_file("/content/drive/MyDrive/GE234/tha_admbnda_adm1_rtsd_20190221.gpkg")

# ตรวจสอบ CRS ของทั้งสอง GeoDataFrame
print(f"CRS of gdf_amphoe: {gdf_amphoe.crs}")
print(f"CRS of gdf_provinces: {gdf_provinces.crs}")

# แก้ไขปัญหา CRS mismatch: reproject gdf_provinces ให้เป็น CRS เดียวกับ gdf_amphoe
gdf_provinces_reprojected = gdf_provinces.to_crs(gdf_amphoe.crs)

# ทำ Spatial Join: เอาข้อมูลจังหวัดไปใส่ในข้อมูลอำเภอที่อยู่ "ข้างใน" (within)
# ใช้ gdf_provinces_reprojected หลังจากการ reproject
joined_data = gpd.sjoin(gdf_amphoe, gdf_provinces_reprojected, how="inner", predicate="within")

# ตรวจสอบคอลัมน์ที่มีอยู่ใน joined_data เพื่อหาชื่อคอลัมน์ที่ถูกต้องสำหรับอำเภอและจังหวัด
print("Columns in joined_data:")
print(joined_data.columns)

# แสดงชื่ออำเภอคู่กับจังหวัดที่ถูก Join กันแล้ว 5 แถวแรก
print(joined_data[['A_NAME_T', 'ADM1_TH']].head())
```

```python
# สมมติเรามีรายการค่า NDVI เฉลี่ยที่คำนวณจากดาวเทียมมาแล้ว (จำนวนต้องเท่ากับจำนวนจังหวัด)
# ในที่นี้ขอกำหนดค่าสุ่มให้แต่ละจังหวัดใน GeoDataFrame
gdf_provinces['avg_ndvi'] = np.random.uniform(0.2, 0.8, len(gdf_provinces))

# หาจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด
top_province = gdf_provinces.loc[gdf_provinces['avg_ndvi'].idxmax()]

print(f"จังหวัดที่มี NDVI เฉลี่ยสูงสุดคือ: {top_province['ADM1_TH']}")
print(f"ค่า NDVI: {top_province['avg_ndvi']:.4f}")
```
